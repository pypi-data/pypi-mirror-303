```python
import climbing_cams as cams
import matplotlib.pyplot as plt
```

```python
c = cams.db.select(name='C4')
m = cams.db.select(name='Master Cam')
t = cams.db.select(brand='Totem')
f = cams.db.select(name='Friend')
```

```python
fig, [ax1, ax2] = plt.subplots(1,2, figsize=(15,5))
fig, ax1 = cams.plots.weight_range(m, ax=ax1, color='orange')
fig, ax1 = cams.plots.weight_range(t, ax=ax1, color='teal')
fig, ax1 = cams.plots.weight_range(f, ax=ax1, color='blue')
ax1.legend()
fig, ax2 = cams.plots.weight_range(c, ax=ax2, color='coral')
fig, ax2 = cams.plots.weight_range(t, ax=ax2, color='teal')
ax2.legend()
plt.show()
```
